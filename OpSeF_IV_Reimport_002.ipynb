# Functions

```python
def splitpath(path, maxdepth=20):
    '''Splits a path in all its parts'''
    ( head, tail ) = os.path.split(path)
    return splitpath(head, maxdepth - 1) + [ tail ] \
        if maxdepth and head and head != path \
        else [ head or tail ]
```

```python
def import_mask_to_img_dics(fp):
    ''' Imports filepair list as dic of dics'''
    # read content
    with open(pair_dic_fn) as f:
        content = f.readlines()
    # remove whitespace characters like `\n` at the end of each line
    content = [x.strip() for x in content]
    # make target dic
    mask_to_img_dic_dic = {}
    # create static part of filepath
    path_as_list = splitpath(fp)
    base_folder = os.path.join(*path_as_list[:-2])
    #print(content[0].split(";"))
    # convert content
    for seg_mask_id in range(1,len(content[0].split(";"))):
        mask_to_img_dic = {}
        if len(content[0].split(";")[seg_mask_id])>1: # not empty...
            for line in content:
                mylist = line.split(";")
                mask_to_img_dic[base_folder+mylist[0]] = base_folder + mylist[seg_mask_id]
            mask_to_img_dic_dic[str(seg_mask_id)] = mask_to_img_dic
    return mask_to_img_dic_dic,base_folder
```
